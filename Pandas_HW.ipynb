```python
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "C:/Users/emhar/Desktop/PythonStuff/schools_complete.csv"
student_data_to_load = "C:/Users/emhar/Desktop/PythonStuff/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data2 = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data2["per student budget"]=school_data2["budget"]/school_data2["size"]
school_data2
```

```python
#average math score for all students
averagem = school_data2["math_score"].mean()
```

```python
#average reading score for all students
averager = school_data2["reading_score"].mean()
```

```python
#counting total students 
total_students = school_data2["Student ID"].count()
```

```python
#finding list of budget values
#what if two districts have the same budget?
unique = school_data2["budget"].unique()
unique
```

```python
#find total budget
total_budget = unique.sum()
total_budget
```

```python
#number of students at each school
count_schools = school_data2["school_name"].value_counts()
```

```python
#number of students that passed math
    #list of students that passed
passing_math = school_data2.loc[school_data2["math_score"] >= 70, ["Student ID"]]
    #length of list
passing_math1=len(passing_math)
```

```python
#percentage of total students that passed math
passing_math2=(passing_math1/total_students)*100
passing_math2
```

```python
#number of students that passed reading 
    #list of students that passed
passing_reading= school_data2.loc[school_data2["reading_score"] >= 70, ["Student ID"]]
    #length of list
passing_reading1=len(passing_reading)
passing_reading1
```

```python
#percentage of total students that passed math
passing_reading2=(passing_reading1/total_students)*100
passing_reading2
```

```python
#overall pass rate
overall_pass=(passing_math2+passing_reading2)/2
overall_pass
```

```python
#create a dataframe including previous calculations
summary_df = pd.DataFrame({
    "Total Schools": [len(count_schools)],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [averagem],
    "Average Reading Score": [averager],
    "% Passing Math": [(passing_math1/total_students)*100],
    "% Passing Reading": [(passing_reading1/total_students)*100],
    "% Overall Passing Rate": [overall_pass]
})

#formatting the dataframe 
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,}".format)
summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Average Math Score"] = summary_df["Average Math Score"].map("{:.2f}".format)
summary_df["Average Reading Score"] = summary_df["Average Reading Score"].map("{:.2f}".format)
summary_df["% Passing Math"] = summary_df["% Passing Math"].map("%{:.2f}".format)
summary_df["% Passing Reading"] = summary_df["% Passing Reading"].map("%{:.2f}".format)
summary_df["% Overall Passing Rate"] = summary_df["% Overall Passing Rate"].map("%{:.2f}".format)
summary_df

#perhaps count math scores and reading scores if you arent sure all students ahve both
```
